Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture

many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

My architecture has four layers with two convolutional and two fully connected layers. The first convolutional layer has 3 input channels and 5 output channels, followed by the ReLu activation function and a pooling layer. The second convolutional layer has 5 input channels and 8 output channels, followed by the ReLu activation function and a pooling layer. The first fully connected layer has 53*53*8 (since 8 output channels)input units and uses the ReLu activation function. The last fully connected layer has 32 input units and 9 output units since we are classifying 9 classes.

```python
  #Convolutional Neural Network Architecture
class CNN_me(nn.Module):
    def __init__(self):
          super(CNN_me, self).__init__()
          self.conv1 = nn.Conv2d(3, 5, 5) #in_channels, out_chanels, kernel_size
          #output = 224-5+1 =220
          #outputunit = 110
          self.conv2 = nn.Conv2d(5, 8, 5) #in_channels, out_chanels, kernel_size
          # = 110-5+1 = 106
          self.pool = nn.MaxPool2d(2, 2) #kernel_size, stride 
          #outputunit = 53
          self.fc1 = nn.Linear(22472, 32) # = 53*53*8
          self.fc2 = nn.Linear(32, 9)

    def forward(self, x):
          x = self.pool(F.relu(self.conv1(x)))
          x = self.pool(F.relu(self.conv2(x)))
          x = x.view(-1, 22472)
          x = F.relu(self.fc1(x))
          x = self.fc2(x)
          return x
```

```python
  #New CNN architecture for hyperprameter tuning part.
class CNN_me_kernel(nn.Module):
    def __init__(self):
          super(CNN_me_kernel, self).__init__()
          self.conv1 = nn.Conv2d(3, 5, 3) #in_channels, out_chanels, kernel_size
          #output = 224-5+1 =220
          # = 220/2 = 110
          self.conv2 = nn.Conv2d(5, 6, 7) #in_channels, out_chanels, kernel_size
          # = 110-7+1 = 104
          self.pool = nn.MaxPool2d(2, 2) #kernel_size, stride 
          #outputunit = 52
          self.fc1 = nn.Linear(16224, 30) # = 52*52*6
          self.fc2 = nn.Linear(30, 9)

    def forward(self, x):
          x = self.pool(F.relu(self.conv1(x)))
          x = self.pool(F.relu(self.conv2(x)))
          x = x.view(-1, 16224)
          x = F.relu(self.fc1(x))
          x = self.fc2(x)
          return x

```

### Part (b) Training Code - 3 pt

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice

```python
# data loading
train_b = datasets.ImageFolder(train_dir, transform = transform)
print('Number of training images:', len(train_b))

loader = torch.utils.data.DataLoader(train_b, batch_size = 64) 
```

```python
def accuracy_metrics(model, loader, use_cuda):
  correct = 0
  n = 0
  for image, label in loader:
    # enable GPU
    if use_cuda and torch.cuda.is_available():
      image = image.cuda()
      label = label.cuda()
    output = model(image)
    # index of the max. prediction score
    pred = output.max(1, keepdim = True)[1]
    correct += pred.eq(label.view_as(pred)).sum().item() 
    n += image.shape[0] 
    
  return correct/n
```

```python
def training_sgd(model, train_data, modelname, val_data, batch_size, num_epochs = 1, lr = 0.01, use_cuda = True):
  # define loaders
  train_loader = torch.utils.data.DataLoader(train_data, batch_size = batch_size, shuffle = True)
  val_loader = torch.utils.data.DataLoader(val_data, batch_size = batch_size, shuffle = True)
  # define loss function used
  criterion = nn.CrossEntropyLoss()
  # define optimizer
  optimizer = optim.SGD(model.parameters(), lr = lr, momentum = 0.9)
  iters, losses, train_acc, val_acc = [], [], [], []
  # training
  i = 0 # track number of iterations
  for epoch in tqdm(range(num_epochs)): # tqdm to track process
    for image, label in train_loader: # for each batch
      # enable GPU usage
      if use_cuda and torch.cuda.is_available():
        model.cuda()
        image = image.cuda()
        label = label.cuda()
        if i == 0:
          print('on GPU')
      output = model(image)           # forward pass
      loss = criterion(output, label) # compute loss
      loss.backward()                 # compute parameter updates
      optimizer.step()                # update parameters
      optimizer.zero_grad()           # clean up step

      # save training information
      if i % 2 == 0:
        iters.append(i)
        losses.append(float(loss)/batch_size) # compute average loss
        # compute training and validation accuracy
        train_acc.append(accuracy_metrics(model, train_loader, use_cuda))
        val_acc.append(accuracy_metrics(model, val_loader, use_cuda))
      i += 1

    # Save the current model (checkpoint) to a file
    path = "model_{0}_bs{1}_lr{2}_epoch{3}".format(modelname, batch_size, lr, epoch)
    model_path = path
    torch.save(model.state_dict(), model_path)
  print('Finished Training')

  # plot training information
  plt.title("Loss vs Iterations")
  plt.plot(iters, losses, label = "Training")
  plt.xlabel("Iterations")
  plt.ylabel("Loss")
  plt.show()
  plt.title("Accuracy vs Iterations")
  plt.plot(iters, train_acc, label = "Training")
  plt.plot(iters, val_acc, label = "Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Accuracy")
  plt.legend()
  plt.show()

  print("Final Training Accuracy: {}".format(train_acc[-1]))
  print("Final Validation Accuracy: {}".format(val_acc[-1]))
  return iters, losses, train_acc, val_acc
```

```python
class Transfer_Learning(nn.Module):
  def __init__(self):
    super(Transfer_Learning, self).__init__()
    self.fc1 = nn.Linear(6*6*256, 32)
    self.fc2 = nn.Linear(32, 9)

  def forward(self, x):
    x = x.view(-1, 6*6*256)
    x = F.relu(self.fc1(x))
    x = self.fc2(x)
    return x
```

```python
def accuracy_metrics_transferL(model, images, labels, use_cuda):
  # enable GPU
    if use_cuda and torch.cuda.is_available():
      images = images.cuda()
      labels = labels.cuda()    
    outputs = model(images)
    prob = F.softmax(outputs)
    
    #select index with maximum prediction score
    pred = prob.max(1, keepdim=True)[1]
    
    # number of correctly classified images
    correct = pred.eq(labels.view_as(pred)).sum().item()
    total = len(images)
    return correct / total
```

```python
def training_with_transfer(model, train_features, train_labels, val_features, val_labels, num_epochs = 1, 
                           lr = 0.01, use_cuda = True):
  criterion = nn.CrossEntropyLoss()
  optimizer = optim.Adam(model.parameters(), lr = lr)
  iters, losses, train_acc, val_acc = [], [], [], []
  
  # training
  n = 0 # track number of iterations
  for epoch in tqdm(range(num_epochs)):
    # enable GPU
    if use_cuda and torch.cuda.is_available():
      model = model.cuda()
      #load the features that pre-trained by alexnet
      train_features = train_features_save.cuda()
      train_labels = train_labels.cuda()
      if n == 0:
        print('on GPU')
    out = model(train_features)         # forward pass
    loss = criterion(out, train_labels) # compute loss
    loss.backward()                     # compute parameter updates
    optimizer.step()                    # update parameters
    optimizer.zero_grad()               # clean up step

    # save training information
    if n % 2 == 0:
      iters.append(n)
      losses.append(float(loss)/len(train_features))                                                   # conpute average loss
      train_acc.append(accuracy_metrics_transferL(model, train_features_save, train_labels, use_cuda)) # compute training accuracy
      val_acc.append(accuracy_metrics_transferL(model, val_features_save, val_labels, use_cuda))       # compute valuation accuracy
    n += 1
    
  # plot training information
  plt.title("Loss vs Iterations")
  plt.plot(iters, losses, label = "Training")
  plt.xlabel("Iterations")
  plt.ylabel("Loss")
  plt.show()
  plt.title("Accuracy vs Iterations")
  plt.plot(iters, train_acc, label = "Training")
  plt.plot(iters, val_acc, label = "Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Accuracy")
  plt.legend()
  plt.show()

  print("Final Training Accuracy: {}".format(train_acc[-1]))
  print("Final Validation Accuracy: {}".format(val_acc[-1]))
  return iters, losses, train_acc, val_acc
```

I designed an architecture that has 2 fully connected layers for the purpose of classification since we have pre-trained CNN model. There's a ReLu activation function applied in the first layer. The first layer has 256x6x6 input neurons and 32 output neurons, the second layer has 32 input neurons and 9 output neurons since we have 9 classes. And we use softmax activation function on the second layer.
